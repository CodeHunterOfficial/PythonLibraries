#Создание и выполнение Python-проекта в VS Code

### 1. Установите Python
1. Скачайте Python с [официального сайта](https://www.python.org/downloads/).
2. Убедитесь, что Python установлен:
   - Откройте терминал (или командную строку).
   - Введите команду:
     ```bash
     python --version
     ```
     или
     ```bash
     python3 --version
     ```
   - Если версия Python отображается, значит, установка прошла успешно.

3. Убедитесь, что Python добавлен в PATH (это обычно происходит автоматически при установке, но лучше проверить).

---

### 2. Установите Visual Studio Code
1. Скачайте и установите VS Code с [официального сайта](https://code.visualstudio.com/).
2. Запустите VS Code.

---

### 3. Установите расширение Python в VS Code
1. Откройте VS Code.
2. Перейдите в раздел расширений (Extensions) через боковую панель или нажмите `Ctrl+Shift+X`.
3. В поисковой строке введите "Python".
4. Найдите расширение от Microsoft и нажмите "Install".

---

### 4. Создайте папку для проекта
1. Создайте папку на вашем компьютере, где будет храниться ваш проект. Например, `my_python_project`.
2. Откройте эту папку в VS Code:
   - В меню выберите `File > Open Folder...` (или нажмите `Ctrl+K Ctrl+O`).
   - Выберите созданную папку.

---

### 5. Создайте виртуальное окружение
1. Откройте терминал в VS Code (`Ctrl+``).
2. Создайте виртуальное окружение:
   ```bash
   python -m venv venv
   ```
   - Здесь `venv` — это имя папки, где будет храниться виртуальное окружение. Вы можете выбрать другое имя, но `venv` — это стандартное название.

---

### 6. Активируйте виртуальное окружение
1. В PowerShell вы можете столкнуться с ошибкой, связанной с политикой выполнения сценариев. Чтобы решить эту проблему:
   - Измените политику выполнения для текущего сеанса:
     ```powershell
     Set-ExecutionPolicy RemoteSigned -Scope Process
     ```
   - После этого активируйте виртуальное окружение:
     ```powershell
     .\venv\Scripts\activate
     ```
   - Если вы не хотите менять политику, используйте командную строку (`cmd`) вместо PowerShell:
     ```cmd
     venv\Scripts\activate
     ```

2. После активации в терминале появится префикс `(venv)`.

---

### 7. Настройте VS Code для использования виртуального окружения
1. Откройте Command Palette (`Ctrl+Shift+P`).
2. Введите и выберите `Python: Select Interpreter`.
3. Выберите интерпретатор из виртуального окружения (например, `venv`).

### 8. Создайте файл Python
1. В боковой панели VS Code нажмите на иконку "New File" (или используйте `Ctrl+N`).
2. Назовите файл, например, `main.py`.
3. Напишите простой код:
   ```python
   print("Hello, World!")
   ```

---

### 9. Установите необходимые пакеты (если нужно)
1. Если ваш проект требует сторонних библиотек, установите их через `pip`. Например:
   ```bash
   pip install requests
   ```
2. Чтобы сохранить список зависимостей, создайте файл `requirements.txt`:
   ```bash
   pip freeze > requirements.txt
   ```

1. Откройте файл `main.py`.
2. Нажмите `F5` или выберите `Run > Start Debugging` для запуска кода.
3. Вы также можете запустить код через терминал:
   ```bash
   python main.py
   ```

### 11. Настройте отладку (опционально)
1. Перейдите в раздел "Run and Debug" (или нажмите `Ctrl+Shift+D`).

3. Выберите "Python" в качестве среды.
4. Теперь вы можете использовать точки останова и другие функции отладки.

---

### 12. Настройте линтеры и форматтеры (опционально)
Для улучшения качества кода установите линтеры и форматтеры, такие как `pylint`, `flake8`, `black` и `autopep8`. Их можно настроить в файле `settings.json` VS Code.

#### Пример `settings.json`:
```json
{
    "python.linting.enabled": true,
    "python.linting.pylintEnabled": true,
    "python.formatting.provider": "black",
    "python.linting.flake8Enabled": true,
    "python.linting.lintOnSave": true,
}
```

#Создание Django-проекта с нуля в VS Code

Django — это мощный фреймворк для создания веб-приложений на Python. В этой лекции мы рассмотрим, как создать Django-проект с нуля, настроить его в Visual Studio Code и запустить локальный сервер.

---

### 1. Установите Python и VS Code
1. Убедитесь, что Python установлен (см. предыдущую лекцию).
2. Установите Visual Studio Code (см. предыдущую лекцию).

---

### 2. Установите Django
1. Откройте терминал в VS Code (`Ctrl+``).
2. Установите Django с помощью `pip`:
   ```bash
   pip install django
   ```
3. Проверьте установку:
   ```bash
   django-admin --version
   ```
   - Если версия Django отображается, установка прошла успешно.

---

### 3. Создайте Django-проект
1. В терминале перейдите в папку, где вы хотите создать проект.
2. Создайте новый Django-проект:
   ```bash
   django-admin startproject myproject
   ```
   - Здесь `myproject` — это название вашего проекта. Вы можете выбрать любое имя.

3. Перейдите в папку проекта:
   ```bash
   cd myproject
   ```

---

### 4. Создайте виртуальное окружение (опционально, но рекомендуется)
1. Создайте виртуальное окружение:
   ```bash
   python -m venv venv
   ```
2. Активируйте виртуальное окружение:
   - На Windows:
     ```bash
     venv\Scripts\activate
     ```
   - На macOS/Linux:
     ```bash
     source venv/bin/activate
     ```

3. Установите Django в виртуальное окружение:
   ```bash
   pip install django
   ```

---

### 5. Настройте VS Code для работы с Django
1. Откройте папку проекта в VS Code:
   - В меню выберите `File > Open Folder...` (или нажмите `Ctrl+K Ctrl+O`).
   - Выберите папку `myproject`.

2. Настройте интерпретатор Python:
   - Откройте Command Palette (`Ctrl+Shift+P`).
   - Введите и выберите `Python: Select Interpreter`.
   - Выберите интерпретатор из виртуального окружения (например, `venv`).

---

### 6. Запустите сервер разработки
1. В терминале выполните команду:
   ```bash
   python manage.py runserver
   ```
2. Откройте браузер и перейдите по адресу `http://127.0.0.1:8000/`.
   - Если вы видите страницу с сообщением "The install worked successfully!", значит, сервер работает.

---

### 7. Создайте Django-приложение
1. В терминале выполните команду:
   ```bash
   python manage.py startapp myapp
   ```
   - Здесь `myapp` — это название вашего приложения. Вы можете выбрать любое имя.

2. Зарегистрируйте приложение в проекте:
   - Откройте файл `myproject/settings.py`.
   - Найдите список `INSTALLED_APPS` и добавьте туда ваше приложение:
     ```python
     INSTALLED_APPS = [
         ...
         'myapp',
     ]
     ```

---

### 8. Создайте представление (view)
1. Откройте файл `myapp/views.py`.
2. Добавьте простое представление:
   ```python
   from django.http import HttpResponse

   def home(request):
       return HttpResponse("Hello, Django!")
   ```

3. Настройте маршрут (URL):
   - Откройте файл `myapp/urls.py` (если его нет, создайте).
   - Добавьте следующий код:
     ```python
     from django.urls import path
     from . import views

     urlpatterns = [
         path('', views.home, name='home'),
     ]
     ```

4. Подключите маршруты приложения к проекту:
   - Откройте файл `myproject/urls.py`.
   - Добавьте следующий код:
     ```python
     from django.contrib import admin
     from django.urls import path, include

     urlpatterns = [
         path('admin/', admin.site.urls),
         path('', include('myapp.urls')),
     ]
     ```

---

### 9. Запустите сервер и проверьте результат
1. В терминале выполните команду:
   ```bash
   python manage.py runserver
   ```
2. Откройте браузер и перейдите по адресу `http://127.0.0.1:8000/`.
   - Если вы видите сообщение "Hello, Django!", значит, всё работает правильно.

---

### 10. Работа с моделями и базой данных
1. Создайте модель в `myapp/models.py`:
   ```python
   from django.db import models

   class Post(models.Model):
       title = models.CharField(max_length=100)
       content = models.TextField()
       created_at = models.DateTimeField(auto_now_add=True)

       def __str__(self):
           return self.title
   ```

2. Создайте и примените миграции:
   ```bash
   python manage.py makemigrations
   python manage.py migrate
   ```

3. Зарегистрируйте модель в админке:
   - Откройте файл `myapp/admin.py`.
   - Добавьте следующий код:
     ```python
     from django.contrib import admin
     from .models import Post

     admin.site.register(Post)
     ```

4. Создайте суперпользователя:
   ```bash
   python manage.py createsuperuser
   ```
   - Следуйте инструкциям на экране.

5. Перейдите в админку:
   - Откройте браузер и перейдите по адресу `http://127.0.0.1:8000/admin/`.
   - Войдите с данными суперпользователя.

---

### 11. Настройка статических файлов
1. Создайте папку `static` в вашем приложении (`myapp/static`).
2. Добавьте CSS, JS или изображения в эту папку.
3. Подключите статические файлы в шаблоне:
   - Откройте файл `myapp/templates/myapp/index.html` (создайте папку `templates/myapp`, если её нет).
   - Добавьте следующий код:
     ```html
     {% load static %}
     <link rel="stylesheet" href="{% static 'myapp/style.css' %}">
     ```

---

### 12. Использование шаблонов (templates)
1. Создайте папку `templates` в вашем приложении (`myapp/templates`).
2. Создайте файл `myapp/templates/myapp/index.html`:
   ```html
   <!DOCTYPE html>
   <html lang="en">
   <head>
       <meta charset="UTF-8">
       <meta name="viewport" content="width=device-width, initial-scale=1.0">
       <title>My Django App</title>
   </head>
   <body>
       <h1>Hello, Django!</h1>
       <p>This is my first Django template.</p>
   </body>
   </html>
   ```

3. Обновите представление в `myapp/views.py`:
   ```python
   from django.shortcuts import render

   def home(request):
       return render(request, 'myapp/index.html')
   ```

---

### 13. Сохраните проект в Git (опционально)
1. Инициализируйте Git-репозиторий:
   ```bash
   git init
   ```

2. Создайте файл `.gitignore` и добавьте туда папку виртуального окружения и другие ненужные файлы:
   ```
   venv/
   __pycache__/
   *.pyc
   *.sqlite3
   ```

3. Добавьте файлы в репозиторий:
   ```bash
   git add .
   git commit -m "Initial commit"
   ```

---

### Итог
Теперь у вас есть полностью настроенный Django-проект с нуля. Вы можете продолжать развивать его, добавляя новые функции, модели, представления и шаблоны. Удачи в разработке! 🚀
